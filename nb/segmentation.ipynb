### Naming Convention:

The model and patch configuration for a given result is expressed in the image's title, which corresponds to the model name.  The model name is written as: ```name_unet_d#_p##_s#_n#```

Where:
- ```name_unet``` specifices the model architecture
- ```d#```  specifies amount of dropout in the model. The number corresponds to digits *after* the decimal point, i.e. d25 would indicate a dropout of 0.25
- ```p#```  specifies the initial patch size that was extracted from the image. 
- ```s#``` specifies the patch's scale, as a percentage. I.e. s050 means the extracted patch was scaled to 50% of it's original size.
- ```n#``` indicates how many patches were extracted from every slice of a NIFTI image.

````python
all_lab = van+rob+dil
def dice_coef(y_true, y_pred,smooth=1):
    ''' Metric used for CNN training'''
    y_true_f = y_true.ravel()
    y_pred_f = y_pred.ravel()
    intersection = sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (sum(y_true_f) + sum(y_pred_f) + smooth)


test_dice = pd.DataFrame(columns=["model","dice","arch","size"])
tru_lab = nib.load("/home/amdanko/projects/def-rfrayne/amdanko/data/maxtree/SNL_acq1_rep1_Sequence_Number_13_0_1_man.nii.gz")
tru_lab = tru_lab.get_data().transpose(2,0,1)  
````

````python
for lab in all_lab:
    pred = np.load(lab)
    dicey = dice_coef(tru_lab,pred)
    test= pd.DataFrame(dict(model=lab.split("/")[-1],arch=lab.split("_")[0].split("/")[1],
                            size=lab.split("_p")[1].split("_")[0],scale=lab.split("_s")[1].split("_")[0],dice=dicey),index=[0])
    test_dice = test_dice.append(test)

test_dice.sort_values("dice",ascending=0)
````

````python
plotty = ["arch","size"]
title = ["Unet", "Patch Size"]

for i in range(len(plotty)):

    a = test_dice.boxplot(column="dice",by=plotty[i],patch_artist=True,return_type="dict",showmeans=True,figsize=(10,10))

    plt.suptitle("")

    # label adjustment
    p = plt.gca()

    p.set_xlabel("")
    p.set_title("DICE per "+title[i], fontsize=25)
    p.tick_params(axis='y', labelsize=15)
    p.tick_params(axis='x', labelsize=20)

    for i in range(3):
        a['dice']['medians'][i].set_linewidth(3)
        a['dice']['caps'][2*i].set_linewidth(3)
        a['dice']['caps'][2*i+1].set_linewidth(3)
        a['dice']['whiskers'][2*i].set_linewidth(2)
        a['dice']['whiskers'][2*i+1].set_linewidth(2)
        a['dice']['medians'][i].set_color('black')
        a['dice']['means'][i].set_markerfacecolor('black')

    colors = ['lightblue', 'lightgreen', 'pink']
    for patch, color in zip(a['dice']['boxes'], colors):
        patch.set_facecolor(color)
    
````

````python
plotty = ["arch","size"]
title = ["Unet", "Patch Size"]

test_dice_s100 = test_dice[test_dice.scale == '100']

for i in range(len(plotty)):

    a = test_dice_s100.boxplot(column="dice",by=plotty[i],patch_artist=True,return_type="dict",showmeans=True,figsize=(10,10))

    plt.suptitle("")

    # label adjustment
    p = plt.gca()

    p.set_xlabel("")
    p.set_title("DICE per "+title[i]+" (scale=100%)", fontsize=25)
    p.tick_params(axis='y', labelsize=15)
    p.tick_params(axis='x', labelsize=20)

    for i in range(3):
        a['dice']['medians'][i].set_linewidth(3)
        a['dice']['caps'][2*i].set_linewidth(3)
        a['dice']['caps'][2*i+1].set_linewidth(3)
        a['dice']['whiskers'][2*i].set_linewidth(2)
        a['dice']['whiskers'][2*i+1].set_linewidth(2)
        a['dice']['medians'][i].set_color('black')
        a['dice']['means'][i].set_markerfacecolor('black')

    colors = ['lightblue', 'lightgreen', 'pink']
    for patch, color in zip(a['dice']['boxes'], colors):
        patch.set_facecolor(color)
    
````
